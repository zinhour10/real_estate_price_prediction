```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('../../../data/processed/land_dataset_road.csv')
```

```python
df.dropna(inplace=True)
```

```python
grouped = df.groupby('h3')['price_per_m2']
df['h3_price_mean'] = df['h3'].map(grouped.mean())
df['h3_price_max'] = df['h3'].map(grouped.max())
df['h3_price_median'] = df['h3'].map(grouped.median())
df['h3_price_min'] = df['h3'].map(grouped.min())
```

```python
df.head()
```

```python
df.drop(columns=['price', 'geometry', 'polygon_geom'], inplace=True)
```

```python
df.head()
```

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns
```

```python
df_encode = pd.get_dummies(df, cat_cols, drop_first=True)
df_encode = df_encode.astype(int)
```
