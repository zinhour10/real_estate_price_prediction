```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_encode = pd.read_csv('../../data/processed/realestates_kh_v7.csv')
```

```python
y = df_encode[['price']]
X = df_encode.drop(columns=['price'])
```

```python
X.describe()
```

```python
import plotly.express as px

cor = X.corr()

fig = px.imshow(
    cor,
    color_continuous_scale='RdBu_r',
    aspect='auto',
    title='Correlation Matrix',
    width=1200,
    height=1200
)
fig.write_html("correlation_matrix.html")
fig.show()
```

```python
cor
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    cor_matrix = dataset.corr()
    for i in range(len(cor_matrix.columns)):
        for j in range(i):
            if abs(cor_matrix.iloc[i, j]) > threshold:  # We are interested in absolute correlation
                colname = cor_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
cor_features = correlation(X, 0.8)
```

```python
len(set(cor_features))
```

```python
cor_features
```

```python
X.drop(columns=cor_features, inplace=True)
```

```python
df_pearson_correlation = pd.concat([X, y], axis=1)
```

```python
df_pearson_correlation.to_csv('../../data/processed/realestates_kh_pearson_correlation_v1.csv', index=False)
```
