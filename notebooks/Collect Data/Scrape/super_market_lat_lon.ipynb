```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Set up the webdriver
options = Options()
options.add_argument("--headless")  # Run in headless mode
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open the Google Maps search URL
initial_url = 'https://www.google.com/maps/search/supermarket+in+Cambodia/@11.5205448,104.864606,11.48z?authuser=0&hl=en&entry=ttu&g_ep=EgoyMDI1MDUyNy4wIKXMDSoASAFQAw%3D%3D'
driver.get(initial_url)
time.sleep(5)  # Allow the page to load

# Scroll to load more locations
scrollable_div = driver.find_element(By.CSS_SELECTOR, 'div[role="feed"]')

last_height = driver.execute_script("return arguments[0].scrollHeight;", scrollable_div)

while True:
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;", scrollable_div)
    time.sleep(5)  # Wait for new data to load

    new_height = driver.execute_script("return arguments[0].scrollHeight;", scrollable_div)
    if new_height == last_height:  # Stop when no new data loads
        break
    last_height = new_height

# Parse the final loaded page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")

# Extract branch details from <a> elements
branch_data = []
branches = soup.find_all("a", class_="hfpxzc")

for branch in branches:
    name = branch.get("aria-label", "").strip()
    url = branch.get("href", "").strip()

    if name and url:
        branch_data.append({"name": name, "url": url})

# Print extracted data
for branch in branch_data:
    print(branch)

# Close the browser
driver.quit()
```

```python
import json # Assuming branch_data might come from a JSON source

result = []

for i in range(len(branch_data)): # Changed to len(branch_data) to include the last element
    temp = {}
    temp['name'] = branch_data[i]['name']

    url = branch_data[i]['url']
    print(f"\n--- Processing index: {i} ---")
    print(f"Original URL: {url}")

    try:
        # Step 1: Split by '/' and get the second-to-last part
        url_parts = url.split('/')
        if len(url_parts) < 2:
            print(f"Error: URL '{url}' does not have enough '/' parts. Skipping.")
            continue # Skip to the next iteration
        relevant_part = url_parts[-2]
        print(f"Relevant URL part (before [1:]): {relevant_part}")

        # Step 2: Remove the leading character (e.g., '1' or 'q=')
        if len(relevant_part) < 2:
            print(f"Error: Relevant part '{relevant_part}' is too short for [1:]. Skipping.")
            continue
        stripped_part = relevant_part[1:]
        print(f"Stripped part (after [1:]): {stripped_part}")

        # Step 3: Split by ','
        coordinates_parts = stripped_part.split(',')
        print(f"Coordinates parts (after split(',')): {coordinates_parts}")

        # Step 4: Take all elements except the last one (assuming it's a name or junk)
        # and ensure there are at least two elements (lat and lon)
        if len(coordinates_parts) < 2:
            print(f"Error: Not enough comma-separated values for lat/lon in '{stripped_part}'. Skipping.")
            continue
        
        # We need to make sure the list has at least 2 elements before trying to access index 0 and 1
        # The `[:-1]` was potentially problematic if it made a list with 2 elements into 1, or 1 into 0.
        # Let's directly try to get lat and lon if available.
        
        lat = coordinates_parts[0]
        lon = coordinates_parts[1]

        # Optional: Add validation if lat/lon are actually numbers
        try:
            float(lat)
            float(lon)
        except ValueError:
            print(f"Warning: Latitude '{lat}' or Longitude '{lon}' are not valid numbers. Skipping.")
            continue

        temp['lat'] = lat
        temp['lon'] = lon
        result.append(temp)

    except IndexError as e:
        print(f"IndexError encountered for branch_data[{i}] (name: {temp['name']}): {e}")
        print(f"Problematic URL: {url}")
        print("This often means the URL format for latitude/longitude was unexpected.")
        # Decide whether to continue, skip, or log the error for later review
        continue # Skip this entry and continue with the next one
    except Exception as e:
        print(f"An unexpected error occurred for branch_data[{i}] (name: {temp['name']}): {e}")
        print(f"Problematic URL: {url}")
        continue # Skip this entry and continue with the next one

print("\n--- Final Result ---")
print(json.dumps(result, indent=2))
```
