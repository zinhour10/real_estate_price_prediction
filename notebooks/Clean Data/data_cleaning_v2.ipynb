```python
# Sort by price_per_m2 (descending or ascending as you prefer)
df = df.sort_values(by='price_per_m2', ascending=True)

# Select the top 5 rows
styled = df.head(1)[['land_area', 'price_per_m2', 'price', 'category_name']].style.set_table_styles({
    'price_per_m2': [{'selector': 'td', 'props': [('min-width', '20px')]}],
    'land_area': [{'selector': 'td', 'props': [('min-width', '20px')]}],
    'price': [{'selector': 'td', 'props': [('min-width', '20px')]}],
    'category_name': [{'selector': 'td', 'props': [('min-width', '20px')]}]
})
styled
```

```python
idx = 3372
# df.at[idx, 'price'] =1200000
df.at[idx, 'land_area'] = 2000
df.at[idx, 'price_per_m2'] = df.at[idx, 'price'] / df.at[idx, 'land_area']
# df.at[idx, 'price'] = df.at[idx, 'land_area'] * df.at[idx, 'price_per_m2']
```

```python
df = df.drop(index=3373)
```

```python
df = df[df['price_per_m2'] >= 25]
```

```python
# Find mean, max, and min of 'price' grouped by 'h_id'
price_stats = df.groupby('h_id')['price_per_m2'].agg(['mean', 'max', 'min']).reset_index()
price_stats
```

```python
# Calculate mean, max, min of 'price_per_m2' grouped by 'h_id'
price_stats = df.groupby('h_id')['price_per_m2'].agg(['mean', 'max', 'min']).reset_index()

# Merge these stats back to the original dataframe on 'h_id'
df = df.merge(price_stats, on='h_id', how='left', suffixes=('', '_price_per_m2'))

# The columns 'mean', 'max', 'min' are now added to df
```

```python
cols = ['mean_per_m2', 'max_per_m2', 'min_per_m2'] 
df.drop(columns=cols, inplace=True)
```

```python
import matplotlib.pyplot as plt
import os
import numpy as np

output_dir = "boxplots_by_hid_batch"
os.makedirs(output_dir, exist_ok=True)

unique_hids = df['h_id'].unique()
batch_size = 20

for i in range(0, len(unique_hids), batch_size):
    batch_hids = unique_hids[i:i+batch_size]
    batch_df = df[df['h_id'].isin(batch_hids)]
    plt.figure(figsize=(18, 6))
    batch_df.boxplot(column='price_per_m2', by='h_id', grid=False, rot=90)
    plt.title(f'Boxplot of price_per_m2 for h_id {i+1} to {i+len(batch_hids)}')
    plt.suptitle('')
    plt.xlabel('h_id')
    plt.ylabel('price_per_m2')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f'boxplot_hid_{i+1}_to_{i+len(batch_hids)}.png'))
    plt.close()
```

```python
import matplotlib.pyplot as plt
import os
import numpy as np

output_dir = "boxplots_by_hid_batch_no_outlier"
os.makedirs(output_dir, exist_ok=True)

def remove_outliers(group):
    Q1 = group['price_per_m2'].quantile(0.25)
    Q3 = group['price_per_m2'].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return group[(group['price_per_m2'] >= lower) & (group['price_per_m2'] <= upper)]

# Remove outliers for each h_id group
df_no_outlier = df.groupby('h_id', group_keys=False).apply(remove_outliers)

unique_hids = df_no_outlier['h_id'].unique()
batch_size = 20

for i in range(0, len(unique_hids), batch_size):
    batch_hids = unique_hids[i:i+batch_size]
    batch_df = df_no_outlier[df_no_outlier['h_id'].isin(batch_hids)]
    plt.figure(figsize=(18, 6))
    batch_df.boxplot(column='price_per_m2', by='h_id', grid=False, rot=90)
    plt.title(f'Boxplot of price_per_m2 (no outlier) for h_id {i+1} to {i+len(batch_hids)}')
    plt.suptitle('')
    plt.xlabel('h_id')
    plt.ylabel('price_per_m2')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f'boxplot_hid_{i+1}_to_{i+len(batch_hids)}_no_outlier.png'))
    plt.close()
```
