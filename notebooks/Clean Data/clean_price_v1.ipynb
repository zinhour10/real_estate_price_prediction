```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/processed/realestates_kh_v2_2.csv')
```

```python
def calculate_price_with_null_handling(row):
    price_str = row['price_display']
    land_size = row['land_area']

    # Handle null or non-string price_display
    if not isinstance(price_str, str):
        return None

    # Remove '$' and ','
    cleaned_str = price_str.replace('$', '').replace(',', '')
    
    # Check if land_size is valid for multiplication
    is_land_size_valid = pd.notna(land_size) and pd.api.types.is_numeric_dtype(type(land_size))

    if '/m¬≤' in cleaned_str or '/m2' in cleaned_str:
        try:
            # Extract price per sqm
            price_per_sqm = float(cleaned_str.replace('/m¬≤', '').replace('/m2', ''))
            
            # Perform multiplication only if land_size is valid
            if is_land_size_valid:
                return price_per_sqm * land_size
            else:
                # If land_size is null/invalid, return None for price per sqm calculations
                return None 
        except ValueError:
            # Handle cases where price per sqm part is not a valid number
            return None
    else:
        try:
            # Directly convert to float for absolute prices
            return float(cleaned_str)
        except ValueError:
            # Handle cases where absolute price is not a valid number
            return None
```

```python
df['price_display'] = df.apply(calculate_price_with_null_handling, axis=1)
```
