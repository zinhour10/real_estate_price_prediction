```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/processed/land_dataset_from_use_lat_lot_getaddr.csv')
```

```python
df = df[df['price_per_m2'] >= 100].copy()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
import folium
from folium.plugins import MarkerCluster
import matplotlib
import matplotlib.pyplot as plt

# Assign a unique color to each h_id
unique_hids = df['h3'].unique()
cmap = plt.get_cmap('tab20', len(unique_hids))  # or use another qualitative colormap
h_id_to_color = {hid: matplotlib.colors.rgb2hex(cmap(i)) for i, hid in enumerate(unique_hids)}

# Center map
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=12)

for _, row in df.iterrows():
    color = h_id_to_color[row['h3']]

    popup_html = f"""
    <b>h_id:</b> {row['h3']}<br>
    <b>Address:</b> {row['ADM3_EN']}<br>
    <b>Area:</b> {row['land_area']:.0f} mÂ²<br>
    <b>location:</b> {row['latitude']},{row['longitude']}<br>
    <b>Price/mÂ²:</b> ${row['price_per_m2']:.2f}
    """

    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=4,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=folium.Popup(popup_html, max_width=300)
    ).add_to(m)

m.save('realestate_price_map_more_info.html')
```
