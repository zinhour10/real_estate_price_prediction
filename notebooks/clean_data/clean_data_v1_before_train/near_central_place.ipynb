```python
# Haversine distance function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    return R * 2 * asin(sqrt(a))
```

```python
df = pd.read_csv("../../../data/raw/mockup/generated_data_10000.csv")
```

```python
# Loop over places and add distance-band columns
for place in places:
    name = place["name"]
    
    def encode_range(row):
        d = haversine(row['latitude'], row['longitude'], place['lat'], place['lon'])
        return {
            f"near_{name}_in_km": int(d),
            f"{name}_nearest": int(d <= 1),
            f"{name}_1_2km": int(1 < d <= 2),
            f"{name}_2_3km": int(2 < d <= 3),
            f"{name}_3_5km": int(3 < d <= 5),
            f"{name}_5_10km": int(5 < d <= 10)
        }

    results = df.apply(encode_range, axis=1, result_type='expand')
    df = pd.concat([df, results], axis=1)
```
