```python
import simplekml

def df_to_kml(df, output_path, lat_col="latitude", lon_col="longitude", name_col="price_per_m2", desc_col="land_area"):
    kml = simplekml.Kml()
    
    for _, row in df.iterrows():
        # Create a new point in KML
        point = kml.newpoint(
            name=row.get(name_col, ""),
            description=row.get(desc_col, ""),
            coords=[(row[lon_col], row[lat_col])]
        )
    
    kml.save(output_path)
    print(f"KML file saved to {output_path}")

# Usage
df_to_kml(df, "locations.kml")
```
