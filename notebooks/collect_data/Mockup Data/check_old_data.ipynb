```python
import pandas as pd
```

```python
old_data = pd.read_csv('../../../data/processed/land_dataset_road.csv')
```

```python
old_data
```

```python
old_data = old_data[~((old_data['price_per_m2'] < 100))]
```

```python
# Define grouping hierarchy (adjust these columns based on your actual data structure)
grouping_columns = ['ADM1_EN', 'ADM2_EN', 'ADM3_EN', 'h3']  # Replace with actual column names

# Aggregate statistics (example: mean and count of price_per_m2)
grouped_stats = old_data.groupby(grouping_columns)['price_per_m2'].agg(
    ['mean', 'median', 'max', 'min']
)
```

```python
import json

# Convert MultiIndex DataFrame to nested dict
def nested_dict_from_multiindex(df):
    result = {}
    for idx, row in df.iterrows():
        d = result
        for key in idx[:-1]:
            d = d.setdefault(key, {})
        d[idx[-1]] = row.to_dict()
    return result

nested_stats = nested_dict_from_multiindex(grouped_stats)
with open('grouped_stats_nested_v2.json', 'w') as f:
    json.dump(nested_stats, f, indent=2)
```
