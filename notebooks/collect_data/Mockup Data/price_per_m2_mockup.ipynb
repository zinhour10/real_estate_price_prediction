```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
df = pd.read_csv('../../../data/processed/mockup_dataset_road.csv')
```

```python
# read sample data
with open('commune_nested_with_min_max.json', 'r') as file:
    data = json.load(file)
```

```python
def get_min_max(row):
    city = row['address_subdivision']
    district = row['address_locality']
    commune = row['address_line_2']
    try:
        commune_list = data[city][district]
        for c in commune_list:
            if c['commune'] == commune:
                return pd.Series({'commune_min': c['min'], 'commune_max': c['max']})
    except KeyError:
        pass
    return pd.Series({'commune_min': None, 'commune_max': None})

# Apply to your DataFrame
df[['commune_min', 'commune_max']] = df.apply(get_min_max, axis=1)
```

```python
distance_columns = ['near_Koh_Pich_in_km', 'near_Russian_Market_in_km', 'near_AEON_Mall_1_in_km', 'near_AEON_Mall_2_in_km', 'near_AEON_Mall_3_in_km', 'near_Bassac_Lane_in_km', 'near_Koh_Norea_in_km', 'near_Camko_City_in_km', 'near_Olympic_Stadium_in_km', 'near_Phsar_Tmey_in_km', 'near_Boeng_Keng_Kang_1_in_km', 'near_Wat_Phnom_in_km', 'near_Chroy_Changvar_Bridge_in_km', 'near_Vattanac_Tower_in_km', 'near_Royal_Palace_in_km', 'near_Sisowath_Riverside_Park_in_km', 'near_Phnom_Penh_Airport_in_km']
amenity_columns = ['n_cafe_5km', 'n_gas_station_5km', 'n_hospital_5km', 'n_hotel_5km', 'n_mart_5km', 'n_pre_school_5km', 'n_secondary_school_5km', 'n_primary_school_5km', 'n_university_5km', 'n_seven_eleven_5km', 'n_resturant_5km', 'n_super_market_5km', 'n_borey_5km', 'n_bank_5km', 'n_atm_5km']
road_columns = ['f_bridleway', 'f_corridor', 'f_cycleway', 'f_disused', 'f_footway', 'f_motorway', 'f_path', 'f_pedestrian', 'f_primary', 'f_residential', 'f_road', 'f_secondary', 'f_service', 'f_steps', 'f_tertiary', 'f_track', 'f_trunk', 'f_trunk_link', 'f_unclassified', 'f_unused']
ROAD_TYPE_WEIGHTS = {
    # High desirability roads (well-maintained, safe, accessible)
    'f_residential': 1.0,    # Residential streets - most desirable
    'f_pedestrian': 0.9,     # Pedestrian zones - very desirable
    'f_cycleway': 0.85,      # Bike paths - desirable for active lifestyle
    'f_footway': 0.8,        # Walking paths - good for accessibility
    
    # Medium desirability roads
    'f_primary': 0.7,        # Major roads - good access but noisy
    'f_secondary': 0.65,     # Secondary roads
    'f_tertiary': 0.6,       # Tertiary roads
    'f_service': 0.55,       # Service roads (alleys, etc.)
    
    # Low desirability roads
    'f_trunk': 0.4,          # Major highways - noisy
    'f_trunk_link': 0.4,     # Highway links
    'f_motorway': 0.3,       # Motorways - very noisy
    'f_unclassified': 0.5,   # Generic roads
    
    # Undesirable roads
    'f_track': 0.2,          # Unpaved tracks
    'f_path': 0.25,          # Generic paths
    'f_steps': 0.3,          # Staircases - inaccessible
    'f_disused': 0.1,        # Disused roads
    'f_unused': 0.1,         # Unused roads
    'f_corridor': 0.15,      # Indoor corridors (uncommon)
    'f_bridleway': 0.2,      # Horse paths
    'f_road': 0.5            # Generic road type
}
```

```python
def calculate_desirability_score(row):
    """Calculate desirability score from 0-1 based on location features"""
    # Inverse distance (closer = higher score)
    distance_score = np.mean([1/(1 + row[col]) for col in distance_columns])
        
    amenity_score = np.mean([np.tanh(row[col]/50) for col in amenity_columns])
    
    # Calculate weighted road score
    road_scores = []
    for col in road_columns:
        if row[col] == 1:  # If this road type is present
            road_scores.append(ROAD_TYPE_WEIGHTS[col])
    
    # If no road types are flagged, use neutral score
    road_score = np.mean(road_scores) if road_scores else 0.5    
    # Combine scores (equal weighting)
    return 0.7 * distance_score + 0.2 * amenity_score + 0.1 * road_score

# Precompute desirability scores
df['desirability_score'] = df.apply(calculate_desirability_score, axis=1)

# Normalize scores within each address_line_2 group
df['desirability_score'] = df.groupby('address_line_2')['desirability_score']\
    .transform(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
def generate_price_per_m2(min_val, max_val, generated_prices, desirability_score, tolerance=0.01):
    """
    Generate price based on 11-tier location desirability with uniqueness constraint
    
    Args:
        min_val: Minimum price value
        max_val: Maximum price value
        generated_prices: Set of already generated prices for this category
        desirability_score: Score from 0-1 (will be mapped to 11 tiers)
        tolerance: Minimum relative difference between prices (default 1%)
    
    Returns:
        Unique price within appropriate band
    """
    # Handle case where min = max
    if min_val == max_val:
        center = min_val
        if center > 1000:
            variation = center * 0.15
        elif center > 100:
            variation = center * 0.20
        else:
            variation = center * 0.30
        min_val = max(0, center - variation)
        max_val = center + variation
    
    # Handle other edge cases
    if max_val <= min_val:
        return round(min_val, 2)
    
    total_range = max_val - min_val
    mean = (min_val + max_val) / 2
    
    # Define 11 price bands with progressively increasing ranges
    if desirability_score >= 0.9:        # Premium tier (90-100%)
        band = (mean + 0.4 * total_range, max_val - 0.02 * total_range)
    elif desirability_score >= 0.8:      # Elite tier (80-90%)
        band = (mean + 0.3 * total_range, mean + 0.4 * total_range)
    elif desirability_score >= 0.7:      # High premium tier (70-80%)
        band = (mean + 0.25 * total_range, mean + 0.3 * total_range)
    elif desirability_score >= 0.6:      # Upper high tier (60-70%)
        band = (mean + 0.2 * total_range, mean + 0.25 * total_range)
    elif desirability_score >= 0.5:      # High tier (50-60%)
        band = (mean + 0.15 * total_range, mean + 0.2 * total_range)
    elif desirability_score >= 0.4:      # Upper middle tier (40-50%)
        band = (mean + 0.1 * total_range, mean + 0.15 * total_range)
    elif desirability_score >= 0.3:      # Middle tier (30-40%)
        band = (mean + 0.05 * total_range, mean + 0.1 * total_range)
    elif desirability_score >= 0.2:      # Lower middle tier (20-30%)
        band = (mean - 0.1 * total_range, mean + 0.05 * total_range)
    elif desirability_score >= 0.1:      # Low tier (10-20%)
        band = (mean - 0.2 * total_range, mean - 0.1 * total_range)
    elif desirability_score > 0:         # Basic tier (0-10%)
        band = (mean - 0.3 * total_range, mean - 0.2 * total_range)
    else:                                # Bottom tier (0%)
        band = (min_val + 0.02 * total_range, mean - 0.3 * total_range)
    
    # Ensure valid range
    low_bound = max(min_val, band[0])
    high_bound = min(max_val, band[1])
    
    # Generate candidate price
    for attempt in range(100):
        candidate = random.uniform(low_bound, high_bound)
        rounded = round(candidate, 2)
        
        # Check uniqueness
        is_unique = True
        for existing in generated_prices:
            if abs(rounded - existing) / total_range < tolerance:
                is_unique = False
                break
                
        if is_unique:
            return rounded
    
    # Fallback to simple random in band
    return round(random.uniform(low_bound, high_bound), 2)
```

```python
# Apply generate_price_per_m2 to each row in df and store results in a new column
generated_prices = set()

def apply_generate_price(row):
    price = generate_price_per_m2(
        row['commune_min'],
        row['commune_max'],
        generated_prices,
        row['desirability_score']
    )
    generated_prices.add(price)
    return price

df['price_per_m2'] = df.apply(apply_generate_price, axis=1)
```
