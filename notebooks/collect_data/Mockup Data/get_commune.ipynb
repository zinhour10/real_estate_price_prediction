```python
import json
```

```python
# read sample data
with open('grouped_stats_nested_v2.json', 'r') as file:
    data = json.load(file)
```

```python
flat_records = []
for subdiv, localities in data.items():
    for locality, streets in localities.items():
        for street in streets.items():
            flat_records.append({
                'city': subdiv,
                'districts': locality,
                'communes': street[0],
            })
```

```python
last_city = None
last_district = None

for record in flat_records:
    city = record['city']
    district = record['district']
    commune = record['commune']
    
    city_str = city if city != last_city else ""
    district_str = district if district != last_district or city != last_city else ""
    
    print(f"{city_str:15} {district_str:20} {commune}")
    
    last_city = city
    last_district = district
```

```python
import json

with open('commune_flat_list.json', 'w', encoding='utf-8') as f:
    json.dump(flat_records, f, ensure_ascii=False, indent=2)
```

```python
import json
from collections import defaultdict

# Load your flat list
with open('commune_flat_list.json', 'r', encoding='utf-8') as f:
    flat_records = json.load(f)

# Build nested dict: {city: {district: [commune, ...]}}
nested = defaultdict(lambda: defaultdict(list))
for rec in flat_records:
    city = rec['city']
    district = rec['district']
    commune = rec['commune']
    # Add as dict with min/max None
    commune_entry = {"commune": commune, "min": None, "max": None}
    # Prevent duplicates
    if not any(c['commune'] == commune for c in nested[city][district]):
        nested[city][district].append(commune_entry)

# Convert defaultdicts to normal dicts for saving as JSON
nested_dict = {city: dict(districts) for city, districts in nested.items()}

# Save to JSON
with open('commune_nested_with_min_max.json', 'w', encoding='utf-8') as f:
    json.dump(nested_dict, f, ensure_ascii=False, indent=2)
```
